```python
# Define a metaclass
class Meta(type):
    
    def __new__(mcs, name, bases, dct):
        print(f'Creating class {name}')
        print(f'Bases: {bases}')
        print(f'Dictionary: {dct}')
        cls = super(Meta, mcs).__new__(mcs, name, bases, dct)
        cls._instance = None
        return cls

    def __init__(cls, name, bases, dct):
        print(f'Initializing class {name}')
        super(Meta, cls).__init__(name, bases, dct)

    def __call__(cls, *args, **kwargs):
        if not cls._instance:
            cls._instance = object.__new__(cls, *args, **kwargs)
        return cls._instance

    def __prepare__(name, bases, **kwargs):
        print(f'Preparing class {name}')
        return super(Meta, Meta).__prepare__(name, bases, **kwargs)

    def __instancecheck__(cls, instance):
        print(f'Checking instance of {cls.__name__}')
        return super(Meta, cls).__instancecheck__(instance)

    def __subclasscheck__(cls, subclass):
        print(f'Checking subclass of {cls.__name__}')
        return super(Meta, cls).__subclasscheck__(subclass)

    def __eq__(cls, other):
        print(f'Comparing {cls.value} with {other.value}')
        return cls.value == other.value
    
class A:
    def __init__(self):
        print('A.__init__ called')

# Define a class using the metaclass
class MyClass(A, metaclass=Meta):
    def __init__(self, value):
        print("Initialized MyClass")
        super().__init__()
        self.value = value

    def display(self):
        print(f'Value: {self.value}')

class B(metaclass=Meta):
    value = 42

    def __init__(self, value):
        self.value = value

    def __eq__(self, other):
        if isinstance(other, B):
            return self.value == other.value
        return False

# Example usage
b1 = B(42)
b2 = B(42)
b3 = B(43)

print(b1 == b2)  # True
print(b1 == b3)  # False

# Create an instance of MyClass
obj = MyClass(10)
print("-------------------")
```

```python
import re
import copy

def resolve_username(usernames, first_name, last_name):
    length = len(last_name)
    original_username = (last_name[:length] + first_name)[:8]
    while length >= 0:
        username = (last_name[:length] + first_name)[:8].lower()
        if username not in usernames:
            if len(username) < len(original_username):
                break
            return username
        length -= 1
    
    counter = 1
    original_username = (last_name[:length] + first_name)[:8]
    while True:
        username = original_username[:8 - len(str(counter))] + str(counter)
        counter += 1
        if username not in usernames:
            return username

def generate_usernames(names):
    usernames = set()  # To track unique usernames
    result = []  # Final list of usernames

    for full_name in names:
        parts = full_name.split()
        given_name = parts[0].lower()
        last_name = parts[-1].lower()

        # Remove invalid characters
        given_name = re.sub(r"[^a-z0-9]", "", given_name)
        last_name = re.sub(r"[^a-z0-9]", "", last_name)

        # Generate base username
        base = (last_name[:7] + given_name)[:8]
        username = base
        # Ensure uniqueness
        counter = 1
        while username in usernames:
            # Try appending numbers, ensuring length is still ≤ 8
            username = resolve_username(usernames, given_name, last_name)
            counter += 1

        # Store unique username
        usernames.add(username)
        result.append(username)

    return result

# Example Usage
names = [
    "Smith John",
    "Smith John",
    "Smith John",
    "Smith John",
    "Smith John",
    "Smith John",
    "Smith John",
    "Smith John",
    "Smith John",
    "Smith John",
    "Smith John",
    "Smith John",
    "Smith John",
    "Smith John",
    "Smith John",
    "John-John Smith",
    "JohnJohn Smith",
    "John Smith"
]

print(generate_usernames(names))
```

## Python generator

```python
def range_python(start, stop, step=1):
    if not isinstance(start, int) or not isinstance(stop, int):
        raise ValueError("Start and end must be integers")
    current = start
    while (current < stop and step > 0) or (current > stop and step < 0):
        yield current
        current += step

for i in range_python(10, 0, -1):
    print(i)
```

```python
def custom_range(start, stop=None, step=1):
    if stop is None:  # If only one argument is provided
        start, stop = 0, start
    if step == 0:
        raise ValueError("step must not be zero")
    current = start
    while (step > 0 and current < stop) or (step < 0 and current > stop):
        yield current
        current += step

# Example usage:
for i in custom_range(1, 10, 2):
    print(i)  # Output: 1, 3, 5, 7, 9
```

## C3 Linearization

```python
class A:
    def method(self):
        print("A's method")

class B(A):
    def method(self):
        print("B's method")
        super().method()

class C(A):
    def method(self):
        print("C's method")
        super().method()

class D(C, B):
    def method(self):
        print("D's method")
        super().method()
        B().method()

# How C3 Linearization Works
# Python's C3 linearization algorithm ensures that:

# - A child class is always checked before its parents
# - If there are multiple parents, they're checked in the order listed in the class definition
# - The algorithm preserves a consistent order that respects both the local precedence (child before parent) and monotonicity (the order of parent classes matters)

d = D()
d.method()
```

### Validators in Python

```python
class Validator:
    """Base descriptor class for attribute validation"""
    def __init__(self, name=None, **kwargs):
        self.name = name
        # Store any additional validation parameters
        for key, value in kwargs.items():
            setattr(self, key, value)
            
    def __set_name__(self, owner, name):
        # Automatically capture the attribute name if not provided
        if self.name is None:
            self.name = name
    
    def __get__(self, instance, owner):
        if instance is None:
            return self
        return instance.__dict__.get(self.name, None)
    
    def validate(self, value):
        """Override in subclasses to implement validation logic"""
        raise NotImplementedError
    
    def __set__(self, instance, value):
        validated_value = self.validate(value)
        instance.__dict__[self.name] = validated_value


class StringField(Validator):
    """Validates that a value is a string with optional length constraints"""
    def __init__(self, name=None, min_length=0, max_length=None, pattern=None, **kwargs):
        super().__init__(name, **kwargs)
        self.min_length = min_length
        self.max_length = max_length
        self.pattern = pattern
        
    def validate(self, value):
        if not isinstance(value, str):
            raise TypeError(f"{self.name} must be a string")
        
        if len(value) < self.min_length:
            raise ValueError(f"{self.name} must be at least {self.min_length} characters")
            
        if self.max_length is not None and len(value) > self.max_length:
            raise ValueError(f"{self.name} cannot exceed {self.max_length} characters")
            
        if self.pattern is not None:
            import re
            if not re.match(self.pattern, value):
                raise ValueError(f"{self.name} does not match the required pattern")
                
        return value


class NumericField(Validator):
    """Validates numeric values with range constraints"""
    def __init__(self, name=None, min_value=None, max_value=None, 
                 allow_negative=True, precision=None, **kwargs):
        super().__init__(name, **kwargs)
        self.min_value = min_value
        self.max_value = max_value
        self.allow_negative = allow_negative
        self.precision = precision
        
    def validate(self, value):
        if not isinstance(value, (int, float)):
            raise TypeError(f"{self.name} must be a number")
            
        if not self.allow_negative and value < 0:
            raise ValueError(f"{self.name} cannot be negative")
            
        if self.min_value is not None and value < self.min_value:
            raise ValueError(f"{self.name} cannot be less than {self.min_value}")
            
        if self.max_value is not None and value > self.max_value:
            raise ValueError(f"{self.name} cannot exceed {self.max_value}")
            
        if self.precision is not None:
            # Round to specified precision
            value = round(value, self.precision)
            
        return value


class DateField(Validator):
    """Validates date values with range constraints"""
    def __init__(self, name=None, min_date=None, max_date=None, format=None, **kwargs):
        super().__init__(name, **kwargs)
        self.min_date = min_date
        self.max_date = max_date
        self.format = format
        
    def validate(self, value):
        from datetime import datetime, date
        
        # Convert string to date if format is provided
        if isinstance(value, str) and self.format:
            try:
                value = datetime.strptime(value, self.format).date()
            except ValueError:
                raise ValueError(f"{self.name} must be in format {self.format}")
                
        if not isinstance(value, (datetime, date)):
            raise TypeError(f"{self.name} must be a date or datetime object")
            
        # Extract date part if datetime
        if isinstance(value, datetime):
            value = value.date()
            
        if self.min_date and value < self.min_date:
            raise ValueError(f"{self.name} cannot be earlier than {self.min_date}")
            
        if self.max_date and value > self.max_date:
            raise ValueError(f"{self.name} cannot be later than {self.max_date}")
            
        return value


# Example usage with financial instruments
class Stock:
    # Type-safe attribute definitions
    ticker = StringField(min_length=1, max_length=5, pattern=r'^[A-Z]+$')
    price = NumericField(min_value=0, precision=2)
    shares_outstanding = NumericField(min_value=1, allow_negative=False)
    last_dividend_date = DateField(format="%Y-%m-%d")
    
    def __init__(self, ticker, price, shares_outstanding=None, last_dividend_date=None):
        self.ticker = ticker
        self.price = price
        self.shares_outstanding = shares_outstanding
        self.last_dividend_date = last_dividend_date
        
    @property
    def market_cap(self):
        if self.shares_outstanding:
            return self.price * self.shares_outstanding
        return None


class Bond:
    isin = StringField(min_length=12, max_length=12, pattern=r'^[A-Z0-9]+$')
    face_value = NumericField(min_value=0)
    coupon_rate = NumericField(min_value=0, max_value=1, precision=4)  # As decimal (0.0525)
    maturity_date = DateField()
    
    def __init__(self, isin, face_value, coupon_rate, maturity_date):
        self.isin = isin
        self.face_value = face_value
        self.coupon_rate = coupon_rate
        self.maturity_date = maturity_date

import datetime
A = Bond('US1234567890', 1000, 0.0525, datetime.datetime.strptime('2025-12-31', '%Y-%m-%d'))
```

```python
import time
import functools
import tracemalloc
import logging
from typing import Any, Callable, Dict, Optional, TypeVar, cast

# Set up logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger('finance_performance')

# Type variables for better type hinting
F = TypeVar('F', bound=Callable[..., Any])

class PerformanceMonitor:
    """
    Decorator class that monitors execution time and memory usage of functions,
    with special handling for exceeding predefined thresholds.
    """
    
    def __init__(
        self,
        time_threshold: float = 1.0,  # in seconds
        memory_threshold: float = 100 * 1024 * 1024,  # 100MB in bytes
        log_level: int = logging.WARNING,
        alert_threshold: float = 5.0,  # Alert if execution time > 5s
        record_args: bool = False,
        custom_callback: Optional[Callable] = None
    ):
        """
        Initialize the performance monitor with thresholds and logging options.
        
        Args:
            time_threshold: Time threshold in seconds
            memory_threshold: Memory threshold in bytes
            log_level: Logging level for threshold violations
            alert_threshold: Time threshold for critical alerts
            record_args: Whether to record function arguments in logs
            custom_callback: Optional callback for threshold violations
        """
        self.time_threshold = time_threshold
        self.memory_threshold = memory_threshold
        self.log_level = log_level
        self.alert_threshold = alert_threshold
        self.record_args = record_args
        self.custom_callback = custom_callback
        
    def __call__(self, func: F) -> F:
        """
        Apply the decorator to the function.
        """
        @functools.wraps(func)
        def wrapper(*args: Any, **kwargs: Any) -> Any:
            # Start memory tracking
            tracemalloc.start()
            start_memory = tracemalloc.get_traced_memory()[0]
            
            # Start timing
            start_time = time.perf_counter()
            
            # Store original function name for nested decorators
            func_name = func.__qualname__
            module_name = func.__module__
            
            try:
                # Execute the function
                result = func(*args, **kwargs)
                return result
            finally:
                # Calculate execution time
                execution_time = time.perf_counter() - start_time
                
                # Calculate memory usage
                current, peak = tracemalloc.get_traced_memory()
                memory_used = peak - start_memory
                tracemalloc.stop()
                
                # Prepare log data
                log_data = {
                    'function': f"{module_name}.{func_name}",
                    'execution_time': execution_time,
                    'memory_used': memory_used,
                    'memory_used_mb': memory_used / (1024 * 1024),
                }
                
                # Add args if configured to do so
                if self.record_args and args:
                    # Only record the first 100 chars of each arg to avoid huge logs
                    log_data['args'] = [str(arg)[:100] + ('...' if len(str(arg)) > 100 else '') 
                                       for arg in args]
                if self.record_args and kwargs:
                    log_data['kwargs'] = {k: str(v)[:100] + ('...' if len(str(v)) > 100 else '') 
                                         for k, v in kwargs.items()}
                
                # Determine if thresholds were exceeded
                time_exceeded = execution_time > self.time_threshold
                memory_exceeded = memory_used > self.memory_threshold
                
                # Log based on threshold violations
                if time_exceeded or memory_exceeded:
                    log_message = f"Performance threshold exceeded: {log_data}"
                    
                    if execution_time > self.alert_threshold:
                        logger.critical(log_message)
                        # Could integrate with alerting systems here
                    else:
                        logger.log(self.log_level, log_message)
                    
                    # Call custom callback if provided
                    if self.custom_callback:
                        self.custom_callback(log_data)
                else:
                    # Debug level for normal execution
                    logger.debug(f"Performance metrics: {log_data}")
        
        return cast(F, wrapper)

# Example of a finance-specific extension of the decorator
def monitor_financial_calculation(
    calculation_type: str = "general",
    critical_for_trading: bool = False,
    **kwargs
) -> Callable[[F], F]:
    """
    Specialized decorator for financial calculations with domain-specific logging.
    
    Args:
        calculation_type: Type of financial calculation (e.g., 'pricing', 'risk')
        critical_for_trading: Whether this calculation is critical for trading decisions
        **kwargs: Additional arguments to pass to PerformanceMonitor
    """
    def alert_handler(log_data: Dict[str, Any]) -> None:
        """Custom handler for financial calculation alerts"""
        if critical_for_trading:
            # In a real system, this might:
            # - Send alerts to a monitoring dashboard
            # - Notify the trading desk
            # - Write to a specialized log for compliance
            logger.critical(
                f"TRADING-CRITICAL calculation performance issue: {log_data}"
            )
    
    # Set appropriate defaults based on calculation type
    if calculation_type == "pricing":
        kwargs.setdefault("time_threshold", 0.5)  # Pricing should be fast
    elif calculation_type == "risk":
        kwargs.setdefault("time_threshold", 2.0)  # Risk calcs can take longer
    
    if critical_for_trading:
        # Critical calculations get stricter thresholds
        kwargs["time_threshold"] = min(kwargs.get("time_threshold", 1.0), 1.0)
        kwargs["log_level"] = logging.ERROR
    
    # Create the decorator with custom alert handler
    return PerformanceMonitor(
        custom_callback=alert_handler,
        **kwargs
    )

# Example usage:

# Basic usage
@PerformanceMonitor(time_threshold=0.1, memory_threshold=10*1024*1024)
def calculate_portfolio_returns(portfolio_data):
    """Calculate returns for a portfolio"""
    time.sleep(0.2)  # Simulate work
    # Create a large array to test memory usage
    large_array = [i for i in range(1000000)]
    return sum(large_array) / len(large_array)

# Domain-specific usage
@monitor_financial_calculation(
    calculation_type="risk",
    critical_for_trading=True,
    record_args=True
)
def calculate_var(portfolio, confidence_level=0.95, time_horizon=1):
    """Calculate Value at Risk for a portfolio"""
    time.sleep(0.3)  # Simulate work
    # ... actual calculation would go here
    return confidence_level * time_horizon * 0.1  # Dummy calculation

# Example with high memory usage
@monitor_financial_calculation(calculation_type="pricing")
def price_exotic_option(underlying_price, volatility, rate, time_to_expiry):
    """Price an exotic option using Monte Carlo simulation"""
    time.sleep(0.2)  # Simulate work
    # Generate large arrays for simulation
    paths = [[i * j for j in range(10000)] for i in range(100)]
    # ... actual calculation would go here
    return underlying_price * 0.1  # Dummy result
```

```python
import random

def quick_sort(arr):
    """Quick Sort Algorithm"""
    if len(arr) <= 1:
        return arr
    pivot = random.choice(arr)
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)


def quick_sort_in_place(arr, low, high):
    """In-place Quick Sort Algorithm"""
    if low < high:
        pi = partition(arr, low, high)
        quick_sort_in_place(arr, low, pi - 1)
        quick_sort_in_place(arr, pi + 1, high)
    return arr

def partition(arr, low, high):
    """Partition function for in-place Quick Sort"""
    pivot = arr[high]
    i = low - 1
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

def merge_sort(arr):
    """Merge Sort Algorithm"""
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left_half = merge_sort(arr[:mid])
    right_half = merge_sort(arr[mid:])
    return merge(left_half, right_half)

def merge(left, right):
    """Merge function for Merge Sort"""
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result
```
